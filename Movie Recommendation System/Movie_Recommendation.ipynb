```python
import ast
def convert(obj):
    m=[]
    for i in ast.literal_eval(obj):
        m.append(i['name'])
    return m
movies_df1['genres']=movies_df1.genres.apply(convert)
movies_df1['keywords']=movies_df1.keywords.apply(convert)
```

```python
def convert_cast(obj):
    count = 0
    l = []
    for i in ast.literal_eval(obj):
        if count != 4:
            l.append(i['name'])
            count += 1
        else:
            break
    return l
movies_df1['cast'] = movies_df1.cast.apply(convert_cast)
```

```python
movies_df.iloc[0].crew
```

```python
def fetch_director(obj):
    import ast # this changes string-list to list
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
movies_df1['director'] = movies_df1.crew.apply(fetch_director)
```

```python
import nltk # reducing the words to their root level
from nltk.stem import PorterStemmer
ps=PorterStemmer()
def stem(text):
    stemmed_words=' '.join([ps.stem(i) for i in text.split()])
    return stemmed_words
movies_df2=movies_df1[['id','title','tag','genres','rating','popularity']]
movies_df2['tag']=movies_df2.tag.apply(lambda x: x.lower())
movies_df2['title']=movies_df2.title.apply(lambda x: x.lower())
movies_df2['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000,stop_words='english')
vectors=tfidf.fit_transform(movies_df2['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie): #item-item based collaborative filtering
    movie=movie.lower()
    movie_index=movies_df2[movies_df2.title==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:7]
    for i in movies_list:
        print(movies_df2.iloc[i[0]].title,' ',movies_df2.iloc[i[0]].genres,' ',movies_df2.iloc[i[0]].rating,' ',movies_df1.iloc[i[0]].original_language)
```

```python
%%writefile app.py
import pickle
import streamlit as st
import requests


st.set_page_config( page_title="Movies Recommendation Engine",
    page_icon=":movie_camera:",
    layout="wide", # Or "centered"
    initial_sidebar_state="collapsed",)
# loading files
with open('movies_df.pkl','rb') as file:
    movies_df=pickle.load(file)

with open('movies_df1.pkl', 'rb') as file:
    movies_df1 = pickle.load(file)

with open('movies_df2.pkl', 'rb') as file:
    movies_df2 = pickle.load(file)

with open('similarity.pkl', 'rb') as file:
    similarity = pickle.load(file)

@st.cache_data
def load_data():
    return movies_df,movies_df1,movies_df2,similarity

movies_df,movies_df1,movies_df2,similarity = load_data()


image_url='https://static.standard.co.uk/2022/11/16/10/netflix-s.jpg?width=1200&height=630&fit=crop'
st.image(image_url)

languages=[]
for i in movies_df.original_language.unique():
  languages.append(i)

language_codes = {
    'en': 'English', 'ja': 'Japanese', 'fr': 'French', 'zh': 'Chinese', 'es': 'Spanish',
    'ko': 'Korean', 'de': 'German', 'hi': 'Hindi', 'ru': 'Russian', 'te': 'Telugu',
    'cn': 'Cantonese', 'it': 'Italian', 'nl': 'Dutch', 'ta': 'Tamil', 'sv': 'Swedish',
    'th': 'Thai', 'da': 'Danish', 'xx': 'Unknown', 'hu': 'Hungarian', 'cs': 'Czech',
    'pt': 'Portuguese', 'is': 'Icelandic', 'tr': 'Turkish', 'nb': 'Norwegian Bokmål',
    'af': 'Afrikaans', 'pl': 'Polish', 'he': 'Hebrew', 'ar': 'Arabic', 'vi': 'Vietnamese',
    'ky': 'Kyrgyz', 'id': 'Indonesian', 'ro': 'Romanian', 'fa': 'Persian', 'no': 'Norwegian',
    'sl': 'Slovenian', 'ps': 'Pashto', 'el': 'Greek'
}

# Replace language codes with their actual names
languages = [language_codes[code] for code in languages]


# Getting Movie Posters
def fetch_poster(movies_id):
  url='https://api.themoviedb.org/3/movie/{}?api_key=4e884d98414cda68e4afb9f25e8c1ee4'.format(movies_id)
  response=requests.get(url)
  response=response.json()
  poster_path=response['poster_path']
  full_path='http://image.tmdb.org/t/p/w500/'+poster_path
  return full_path

def top_10movies(df,top_n): #demographic filtering
  top_movies_posters=[]
  sorted_df=df.sort_values(by=['popularity','vote_count'],ascending=False)
  top_movies=sorted_df.head(top_n)
  top_movies_id=top_movies['id'].tolist()
  top_movies_titles = top_movies['title'].tolist()
  top_movies_genres = top_movies['genres'].tolist()
  top_movies_rating = top_movies['rating'].tolist()
  for i in top_movies_id:
    top_movies_posters.append(fetch_poster(i))
  top_movies_genres = [', '.join(genres).capitalize() for genres in top_movies_genres]
  return top_movies_titles,top_movies_genres,top_movies_rating,top_movies_posters





st.markdown("<h1>Popular Movies</h1>", unsafe_allow_html=True)
st.write("---")  # Horizontal
st.markdown("<div style='overflow-x: auto; white-space: nowrap;'>", unsafe_allow_html=True)
top_movies_titles, top_movies_genres, top_movies_rating, top_movies_posters = top_10movies(movies_df1, 10)
cols = st.columns(5)
for i, col in enumerate(cols):
    with col:
        st.image(top_movies_posters[i], width=200)
        st.markdown(f"<p style='font-size: 20px;'><b>{top_movies_titles[i]}</b></p>", unsafe_allow_html=True)
        st.markdown(f"<p style='font-size: 18px;'><b>Genres: {top_movies_genres[i]}</b></p>", unsafe_allow_html=True)
        st.markdown(f"<p style='font-size: 18px;'><b>Rating: {top_movies_rating[i]}</b></p>", unsafe_allow_html=True)
for i, col in enumerate(cols):
    with col:
        idx=i+5
        st.image(top_movies_posters[idx], width=200)
        st.markdown(f"<p style='font-size: 20px;'><b>{top_movies_titles[idx]}</b></p>", unsafe_allow_html=True)
        st.markdown(f"<p style='font-size: 18px;'><b>Genres: {top_movies_genres[idx]}</b></p>", unsafe_allow_html=True)
        st.markdown(f"<p style='font-size: 18px;'><b>Rating: {top_movies_rating[idx]}</b></p>", unsafe_allow_html=True)


# filtering by language
selected_language=st.selectbox('select language',['select language']+sorted(languages))
movie_title=[]
def filter_by_language(selected_language):
  if selected_language!='select language':
    language_codes_reverse = {v: k for k, v in language_codes.items()}
    selected_language_code = [language_codes_reverse[selected_language]]
    for index, row in movies_df1.iterrows():
      if row['original_language'] == selected_language_code:
        movie_title.append(row['title'])
  return movie_title
movie_titles = filter_by_language(selected_language)

# filtering by title
if movie_titles:
  movie_titles=['select a movie']+ movie_titles
  movie=st.selectbox('Select movie', movie_titles)
  def filter_by_title(movie):
    movie_overview = ''
    movie_genre = ''
    movie_rating = ''
    movie_id = ''
    movie_poster = ''
    for index, row in movies_df1.iterrows():
      if row['title'] == movie:
        movie_overview = row['overview']
        movie_overview = ' '.join(movie_overview).capitalize()
        movie_genre = row['genres']
        movie_genre=', '.join(movie_genre).capitalize()
        movie_rating = row['rating']
        movie_id = row['id']
        movie_poster = fetch_poster(movie_id)
        break  # Stop iterating once the movie is found
    return movie_overview, movie_genre, movie_rating, movie_poster
  movie_overview, movie_genre, movie_rating, movie_poster= filter_by_title(movie)
  col=st.columns(1)
  if movie_poster:
    st.image(movie_poster, width=200)
    st.markdown(f"<p style='font-size: 20px;'><b><u>Story Line</u></b>: <b>{movie_overview}</b></p>", unsafe_allow_html=True)
    st.markdown(f"<p style='font-size: 18px;'><b>Genre: {movie_genre}</b></p>", unsafe_allow_html=True)
    st.markdown(f"<p style='font-size: 18px;'><b>Rating: {movie_rating}</b></p>", unsafe_allow_html=True)

else:
  st.warning('No movies available for the selected langauge')

#Recommendations based on item similarities
def recommend(movie):
    movie=movie.lower()
    movie_index=movies_df2[movies_df2.title==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:7]
    recommended_movies=[]
    recommended_posters=[]
    recommended_genres=[]
    rating=[]
    for i in movies_list:
        movie_titles= movies_df2.iloc[i[0]].title
        recommended_movies.append(movie_titles)
        movies_id=movies_df2.iloc[i[0]].id
        recommended_posters.append(fetch_poster(movies_id))
        recommended_movies = [movie.capitalize() for movie in recommended_movies]
        genres= movies_df2.iloc[i[0]].genres
        genres_str = ', '.join(genres).capitalize() if genres else ''
        ratings= movies_df2.iloc[i[0]].rating
        recommended_genres.append(genres_str)
        rating.append(ratings)
    return recommended_movies,recommended_posters,recommended_genres,rating
if movie_titles:
  if movie!='select a movie':
    if st.button('Also Similar Movies'):
      recommended_movies, recommended_posters, recommended_genres, rating = recommend(movie)
      cols = st.columns(6)
      for i, col in enumerate(cols):
          with col:
              st.image(recommended_posters[i], width=200)
              st.markdown(f"<p style='font-size: 20px;'><b>{recommended_movies[i]}</b></p>", unsafe_allow_html=True)
              st.markdown(f"<p style='font-size: 18px;'><b>Genres: {recommended_genres[i]}</b></p>", unsafe_allow_html=True)
              st.markdown(f"<p style='font-size: 18px;'><b>Rating: {rating[i]}</b></p>", unsafe_allow_html=True)
```
